```python
from pptx import Presentation
from pptx.util import Inches

# PowerPoint 프레젠테이션 작성
prs = Presentation()

# 슬라이드 레이아웃 선택
title_slide_layout = prs.slide_layouts[0]
title_body_slide_layout = prs.slide_layouts[1]
body_slide_layout = prs.slide_layouts[6]
table_slide_layout = prs.slide_layouts[5]

# 제목 슬라이드 추가
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "기획서: 신상품의 개발"
subtitle.text = "작성일: 2023년 11월 10일"

# 슬라이드 2 추가 (개요)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "개요"
body.text = "당사는 현재 인기 상품을 대신할 신상품의 개발을 기획하고 있습니다. 기존의 상품과는 다른 디자인과 기능을 갖추고 새로운 고객층을 확보하는 것이 목표입니다."

# 슬라이드 3 추가 (목적)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "목적"
body.text = "이 기획의 목적은 다음과 같습니다.\n\n- 기존 상품에 싫증을 느끼는 고객층 확보\n- 신상품 판매에 따른 매상 증가\n- 당사 브랜드 가치 향상"

# 슬라이드 4 추가 (계획)
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
title.text = "계획"

# 슬라이드 5 추가 (상품 개발)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "상품 개발"
body.text = "1. 신상품 컨셉 결정\n2. 디자인 결정\n3. 기능 결정\n4. 원재료 조달"

# 슬라이드 6 추가 (시제품 개발)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "시제품 개발"
body.text = "1. 3D 프린터를 사용한 시제품 제작\n2. 시제품 테스트\n3. 문제점 수정"

# 슬라이드 7 추가 (상품 제조)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "상품 제조"
body.text = "1. 생산 라인 설계\n2. 제조 과정 결정\n3. 제조 공장 설립"

# 슬라이드 8 추가 (판매)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "판매"
body.text = "1. 판매 전략 결정\n2. 마케팅 실시\n3. 판매 채널 확립"

# 슬라이드 9 추가 (예산)
slide = prs.slides.add_slide(table_slide_layout)
title = slide.shapes.title
title.text = "예산"
table = slide.shapes.add_table(rows=5, cols=2, left=Inches(0.5), top=Inches(1.5), width=Inches(8.0), height=Inches(1.0)).table
table.cell(0, 0).text = "항목"
table.cell(0, 1).text = "금액"
table.cell(1, 0).text = "상품 개발 비용"
table.cell(1, 1).text = "100,000,000원"
table.cell(2, 0).text = "시제품 개발 비용"
table.cell(2, 1).text = "50,000,000원"
table.cell(3, 0).text = "상품 제조 비용"
table.cell(3, 1).text = "500,000,000원"
table.cell(4, 0).text = "마케팅 비용"
table.cell(4, 1).text = "200,000,000원"

# 슬라이드 10 추가 (일정)
slide = prs.slides.add_slide(table_slide_layout)
title = slide.shapes.title
title.text = "일정"
table = slide.shapes.add_table(rows=5, cols=2, left=Inches(0.5), top=Inches(1.5), width=Inches(8.0), height=Inches(1.0)).table
table.cell(0, 0).text = "항목"
table.cell(0, 1).text = "기간"
table.cell(1, 0).text = "상품 개발"
table.cell(1, 1).text = "3개월"
table.cell(2, 0).text = "시제품 개발"
table.cell(2, 1).text = "1개월"
table.cell(3, 0).text = "상품 제조"
table.cell(3, 1).text = "6개월"
table.cell(4, 0).text = "마케팅"
table.cell(4, 1).text = "2개월"

# 파일로 저장
prs.save("plan.pptx")
```
